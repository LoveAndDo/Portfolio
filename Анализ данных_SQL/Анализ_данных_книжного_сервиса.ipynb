## Изучение обзоров и оценок

**Определим количество обзоров и среднюю оценку для каждой книги.**

```python
query = '''
    SELECT 
        books.book_id,
        books.title,
        COUNT(DISTINCT reviews.review_id) AS review_count,
        AVG(ratings.rating) AS avg_rating
    FROM 
        books
    LEFT JOIN ratings ON books.book_id = ratings.book_id
    LEFT JOIN reviews ON books.book_id = reviews.book_id
    GROUP BY 
        books.book_id, books.title
    ORDER BY 
        review_count DESC;
   
   '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

## Определение ключевых издательств

**Найдем издательство, которое выпустило наибольшее количество книг толще 50 страниц. Это поможет исключить брошюры и сосредоточиться на более значимых изданиях.**

```python
query = '''
    SELECT 
        publishers.publisher_id AS publisher_id,
        publishers.publisher AS publisher_name,
        COUNT(books.book_id) AS book_count
    FROM 
        publishers
    LEFT JOIN books ON publishers.publisher_id = books.publisher_id AND books.num_pages > 50
    GROUP BY 
        publishers.publisher_id, publishers.publisher
    ORDER BY 
        book_count DESC
    LIMIT 1;

   '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

1) <u>Лидер рынка:</u>

## Идентификация выдающихся авторов 

**Определим автора с самой высокой средней оценкой книг, учитывая только книги с 50 и более оценками. Это позволит выявить наиболее ценные для читателей авторы.**

```python
query = '''
    SELECT
        authors.author_id,
        authors.author,
        AVG(sub.avg_rating) AS final_avg_rating
    FROM
        authors
    JOIN (
        SELECT
            books.author_id,
            AVG(ratings.rating) AS avg_rating
        FROM
            books
        JOIN ratings ON books.book_id = ratings.book_id
        GROUP BY
            books.book_id, books.author_id
        HAVING
            COUNT(ratings.rating) >= 50
        ) AS sub ON authors.author_id = sub.author_id
    GROUP BY
        authors.author_id, authors.author
    ORDER BY
        final_avg_rating DESC
    LIMIT 1;
   
   '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

##  Анализ активности пользователей 

**Посчитаем среднее количество обзоров от пользователей, которые поставили больше 48 оценок. Это даст понимание о самых активных и вовлеченных пользователях.**

```python
query = '''
   SELECT
        AVG(sub.review_count) AS avg_review_count
    FROM (
        SELECT
            username,
            COUNT(*) AS review_count
        FROM
            reviews
        WHERE
            username IN (
                SELECT
                    username
                FROM
                    ratings
                GROUP BY
                    username
                HAVING
                    COUNT(*) > 48
            )
        GROUP BY
            username
    ) AS sub;
   
   '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```
